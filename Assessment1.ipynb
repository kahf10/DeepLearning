```python
from PIL import Image

img = Image.open(filepath)

# Get the size of the image
width, height = img.size

# Define the coordinates of the rectangle
x1 = width - 30
y1 = 0
x2 = width
y2 = 30

# Create a black rectangle and paste it onto the original image
black_rectangle = Image.new("RGB", (30, 30), (0, 0, 0))
img.paste(black_rectangle, (x1, y1, x2, y2))

# Save the modified image
img.save("image.jpg")
```
